```python
import math
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Suicide data

```python
# read in file
df_s = pd.read_csv("suicide.csv")
#drop na's
df_s2 = df_s.dropna()
#filter to only us data
df_US = df_s2[df_s2["country"]=='United States']
df_USmean = df_US.groupby('year').mean()
df_USmedian = df_US.groupby('year').median()
df_ge = df_US.groupby
# groupby sex
df_sex = df_US.groupby("sex").mean()
df_sex1 = df_US.groupby("sex").median()
# drop all colmuns besides suicide number
df_sex2 = df_sex.drop(['year', 'population', 'suicides/100k pop', 'HDI for year', 'gdp_per_capita ($)' ], axis = 1)
df_sex3 = df_sex1.drop(['year', 'population', 'suicides/100k pop', 'HDI for year', 'gdp_per_capita ($)' ], axis = 1)
#plot mean and median count of suicide by sex
df_sex2.plot(kind = 'bar')
plt.title("Average Suicide count by Sex")
plt.ylabel("Count")
df_sex3.plot(kind = 'bar')
plt.title("Median Suicide count by Sex")
plt.ylabel("Count")
plt.show()
```

```python
# plot mean and median suicides over the years
df_USmean.plot(y = 'suicides_no')
plt.title("Mean Suicides in US")
df_USmedian.plot( y = 'suicides_no')
plt.title("Median Suicides in US")
plt.show()
df_sex
```
